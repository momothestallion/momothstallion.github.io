```python
#Replacing null values with the specified values
movies_df['audience_rating'].fillna(movies_df['audience_rating'].mean(),inplace=True)
movies_df['audience_count'].fillna(movies_df['audience_count'].mean(),inplace=True)
```

```python
movies_df['in_theaters_date'].fillna(movies_df['on_streaming_date'],inplace=True)
```

```python
movies_df.isnull().sum()
```

```python
#Replacing null values with the value from the previous row/column
movies_df['genre'].ffill(inplace=True)
```

```python
movies_df.isnull().sum()
```

```python
rating= movies_df.rating.value_counts()
rating
```

```python
movies_df['rating'].replace(['PG-13)', 'R)'],['PG-13','R'],inplace=True)
```

```python
rating= movies_df.rating.value_counts().sort_values(ascending=False)
rating
```

```python
plt.figure(figsize=(20,10))
plt.title('rating distribution')
plt.pie(rating,labels=rating.index, autopct='%1.1f%%', startangle=180, shadow=False)
```

```python
movies_df.in_theaters_date =pd.to_datetime(movies_df['in_theaters_date'])
```

```python
movies_df.on_streaming_date =pd.to_datetime(movies_df['on_streaming_date'])
```

```python
movies_df['release_year_month'] =movies_df['in_theaters_date'].dt.to_period('M')
movies_df
```

```python
abc=movies_df.groupby('release_year_month')['release_year_month'].count().sort_values(ascending=False)
abc
```

```python
movies_df.groupby('release_year_month')['release_year_month'].count().plot(kind='line', figsize=(20,10))
```

```python
#compare average tomatometer ratings by studio. which studios produce the highest rated films, on average? the lowest?
addy=movies_df.groupby('studio_name')['tomatometer_rating'].mean().sort_values(ascending= False).reset_index()
addy
```

```python
movies_df['compare'] = abs(movies_df['tomatometer_rating'] - movies_df['audience_rating'])
movies_df
```

```python
cfo=movies_df.groupby(['in_theaters_date','movie_title'])['compare'].sum().sort_values(ascending= False)
cfo
```
